```python
import threading
import requests
from bs4 import BeautifulSoup
from lxml import etree

class WebScraper:
    threads = {}  # Attribut de classe pour stocker les threads
    results = {}  # Attribut de classe pour stocker les résultats

    def __init__(self, url, thread_name):
        self.url = url
        self.thread_name = thread_name
        WebScraper.threads[thread_name] = threading.Thread(target=WebScraper.fetch, args=(url, thread_name))

    @classmethod
    def fetch(cls, url, thread_name):
        try:
            print(f"{thread_name} is fetching {url}")
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            tree = etree.HTML(str(soup))

            # Extraction des balises et attributs avec XPath
            element = tree.xpath("//a/@href")[0] if tree.xpath("//a/@href") else "No link found"
            cls.results[thread_name] = element

            print(f"{thread_name} has finished fetching {url}")
        except Exception as e:
            print(f"Error in {thread_name}: {e}")
            cls.results[thread_name] = None

    @classmethod
    def start(cls):
        for thread in cls.threads.values():
            thread.start()
        for thread in cls.threads.values():
            thread.join()  # Assure que tous les threads se terminent avant de continuer

# Liste d'URLs à scraper
urls = [
    "https://fr.wikipedia.org/wiki/Wikip%C3%A9dia:Accueil_principal",
    "https://fr.wikipedia.org/wiki/Rhinella_marina",
    "https://fr.wikipedia.org/wiki/Amphibia",
    "https://fr.wikipedia.org/wiki/Batrachia"
]

for i, url in enumerate(urls):
    WebScraper(url, f"frog{i}")

print("Before starting threads:", WebScraper.threads)

WebScraper.start()

print("After threads execution, results:", WebScraper.results)
```
